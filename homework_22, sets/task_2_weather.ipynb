```python
days = 30
raining = 12
windy = 8
cold = 4
```

raining ∩ windy ∩ cold = 1
raining ∩ windy = 5
raining ∩ cold = 3
windy ∩ cold = 2

```python
#days, with 2 different weather
raining_windy_cold = 1
r_w = 5 - raining_windy_cold
r_c = 3 - raining_windy_cold
w_c = 2 - raining_windy_cold

days_2_weather = r_w + r_c + w_c
print(days_2_weather)
```

```python
#days, with same weather
rain_only = raining - r_w - r_c - raining_windy_cold
cold_only = cold - r_c - w_c - raining_windy_cold
windy_only = windy - w_c - r_w - raining_windy_cold

days_1_weather = rain_only + cold_only + windy_only
print(days_1_weather)
```

```python
sunny_days = days - days_2_weather - days_1_weather - raining_windy_cold
print(sunny_days)
```
