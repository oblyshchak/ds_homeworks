```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel('cars.xlsx')
data
```

```python
cost_dollar = 40
data['cost in dollars'] = (data['cost'] / cost_dollar).round(2)
```

```python
sort_frame = data.sort_values(by=['fuel consumption', 'cost in dollars'])
sort_frame
```

```python
fid, ax = plt.subplots()
plt.plot(sort_frame['fuel consumption'], sort_frame['cost in dollars'], color = 'blue', linewidth = 3)
plt.xlabel("fuel consumption, l")
plt.ylabel("cost, $")
```

```python
plt.scatter(sort_frame['fuel consumption'], sort_frame['max speed'], color = 'red', marker = '^')
plt.xlabel("fuel consumption, l")
plt.ylabel("max speed, km/h")
```

```python
plt.scatter(sort_frame['carrying capacity'], sort_frame['max speed'], color = 'orange', marker = 'o')
plt.xlabel("carrying capacity, kg")
plt.ylabel("max speed, km/h")
```

```python
plt.hist(sort_frame['max speed'], 5, color = 'purple')
plt.ylabel("frequency")
plt.xlabel("max speed, km/h")
```

```python
sns.boxplot(x='car', y = 'max speed', data = sort_frame)
```

```python
df = sort_frame.drop('cost', axis=1)
sns.set_palette("BuGn_r")

sns.pairplot(df, height=1.5)
```
