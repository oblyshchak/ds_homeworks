```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
data = pd.read_csv('communities.csv')
pd.set_option('display.max_columns', None)
data
```

```python
def error_columns(sign):
    error_columns_list = []
    for column in data.columns:
        try:
            print(f"{column} {data[column].value_counts()[sign]}")
            error_columns_list.append(column)
        except:
            pass
    return error_columns_list
```

```python
error = error_columns('?')
```

```python
data = data.replace(['?'], np.nan)
```

```python
data.head()
```

```python
def convert_type(name_column):
    global data
    convert_dict = {name_column : float
                }
    data = data.astype(convert_dict)
```

```python
for name in error:
    convert_type(name)
```

```python
data.dtypes
```

```python
data = data.interpolate(method = 'linear')
```

```python
data['0.36'].isna().unique()
```

```python
data_clean = data.dropna(axis='columns')
data_clean.head()
```

```python
y = data_clean.iloc[:, -1]
```

```python
x = data_clean.drop(['0.2.2', 'Lakewoodcity'], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.7, random_state = 1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
def metrics_model(test, pred):
    print(f"score for LR model {r2_score(test, pred)}")
    print(f"mae for LR model {mean_absolute_error(test, pred)}")
    print(f"rmse for LR model {np.sqrt(mean_squared_error(test, pred))}")
```

```python
metrics_model(y_test,  y_pred)
```

#Lasso

```python
lasspreg = Lasso(alpha = 0.01)
lasspreg.fit(x_train, y_train)
y_pred_lasso = lasspreg.predict(x_test)
```

```python
metrics_model(y_test,  y_pred_lasso)
```

#Ridge

```python
ridgpreg = Ridge(alpha = 1)
ridgpreg.fit(x_train, y_train)
y_pred_ridg = ridgpreg.predict(x_test)
```

```python
metrics_model(y_test,  y_pred_ridg)
```

#Lasso + Ridge

```python
elastic = ElasticNet(alpha = 0.01, l1_ratio= 1)
elastic.fit(x_train, y_train)
y_pred_elastic = elastic.predict(x_test)
```

```python
metrics_model(y_test,  y_pred_elastic)
```
