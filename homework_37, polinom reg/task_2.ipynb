```python
def error_columns(sign):
    error_columns_list = []
    for column in data.columns:
        try:
            print(f"{column} {data[column].value_counts()[sign]}")
            error_columns_list.append(column)
        except:
            pass
    return error_columns_list
```

```python
error = error_columns('?')
print(error)
```

```python
data = data.replace(['?'], np.nan)
```
